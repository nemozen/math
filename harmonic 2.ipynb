```python
%matplotlib inline
from math import sin, pi
import pandas
import matplotlib.pyplot as plt


def indic(t, k):
    if (t + k/2 + 1) % k == 0:
        return 1
    return 0


def product(l):
    p = 1
    for n in l:
        p *= n
    return p


def harmonic_plot(x, T):
    """Plots v and v2, which is just v time-shifted by
    T.  So if it's periodic they should overlap."""
    df = pandas.DataFrame(data=[
        {'t': t, 
         'v': product(sin(2*pi*(t/100.0)/k) for k in x),
         'v2': product(sin(2*pi*(T+t/100.0)/k) for k in x) 
        } 
        for t in range(0,100*T)])

    fig, axes = plt.subplots(1, 1)
    df.plot(x='t', y=['v', 'v2'], ax=axes, figsize=(10, 8))
    plt.grid()
```

```python
# solution for harmonic sum = 2 with 11
harmonic_plot([1, 5, 6, 8, 9, 10, 12, 15, 18, 20, 24], 24)
harmonic_plot([1, 5, 6, 8, 9, 10, 12, 15, 18, 20, 24], 90)
harmonic_plot([1, 5, 6, 8, 9, 10, 12, 15, 18, 20, 24], 180)
```

The two lines overlap perfectly when T=180, which is the period.

```python
# solution for harmonic sum = 2 with 8
harmonic_plot((1, 3, 4, 9, 10, 12, 15, 18), 45)
harmonic_plot((1, 3, 4, 9, 10, 12, 15, 18), 90)
```

This one is pretty.... Within each period, the two halves are mirror images.
