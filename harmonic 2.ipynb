```python
import pandas
from math import sin, pi

x = [1, 5, 6, 8, 9, 10, 12, 15, 18, 20, 24]
T=720

def indic(t, k):
    if (t + k/2 + 1) % k == 0:
        return 1
    return 0

df = pandas.DataFrame(data=[
    {'t': t, 
     'v1': sum(sin(2*pi*(t/100.0)/k) for k in x),
     'v2': sum(sin(2*pi*(T+t/100.0)/k) for k in x),
     'g':  sum(indic(t/100, k) for k in x),
    } 
    for t in range(0,100*T)])
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 1)
df.plot(x='t', y=['v1', 'v2', 'g'], ax=axes, figsize=(64, 48))
plt.grid()
```

```python
print(df['g'].mean())
```
